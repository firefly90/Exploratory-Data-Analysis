```python
def speech():
    l=[]
    url='https://www.news18.com/news/india/independence-day-full-text-of-narendra-modis-red-fort-speech-2271575.html'
    resp=requests.get(url)
    if resp.status_code==200:
        print('Successfully opened the page')
        print('The speech is as below:\n')
        soup=BeautifulSoup(resp.text,'html.parser')
        table = soup.findAll('div',attrs={"class":"hideCont clearfix paragraph"})
    for x in table:
        r=x.find_all('p')
        for rows in r:
            l.append(rows.get_text())
    return l
        
```

```python
l=speech()
```
