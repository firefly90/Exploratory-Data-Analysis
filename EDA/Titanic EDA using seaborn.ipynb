```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.info()
```

```python
titanic.head()
```

```python
sns.jointplot(data=titanic,y='age',x='fare',kind='scatter',color='orange')
```

```python
sns.boxplot(data=titanic,y='age',x='class')
```

```python
sns.swarmplot(data=titanic,x='class',y='age',palette='Set1')
```

```python
sns.countplot(titanic['sex'])
```

```python
plt.figure(figsize=(12,10))
sns.set_context('talk',font_scale=.8)
sns.heatmap(titanic.corr(),cmap='coolwarm',annot=True)
```

```python
sns.FacetGrid
```

```python
g=sns.FacetGrid(data=titanic,col='sex')
g.map(plt.hist,'age')
```
